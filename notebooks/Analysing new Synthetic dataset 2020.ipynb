```python
import os
import yaml
def get_config(filepath: str = 'config.yml'):
    path = os.getcwd()
    if 'notebook' in path:
        path = os.path.dirname(path)

    with open(os.path.join(path, filepath), 'r') as f:
        config = yaml.safe_load(f.read())
    return config

def read_ground_truth(generator_path: str):
    tsgen_config = get_config('config.yaml')
    ground_truth = pd.DataFrame(tsgen_config['params']['transition_map'], columns=['row', 'length', 'to_model'])
    for col in ground_truth.columns:
        ground_truth[col] = ground_truth[col].astype(int)
    ground_truth['from_model'] = ground_truth.to_model.shift(1).fillna(1).astype(int)
    ground_truth['from_row'] = ground_truth.row.shift(1).fillna(0).astype(int)
    ground_truth['row_end_switch'] = ground_truth.row + ground_truth.length
    ground_truth['from_model_range_plus_target'] = list(zip(zip(ground_truth.from_row, ground_truth.row), ground_truth.from_model))
    return ground_truth[['from_row', 'row', 'length', 'row_end_switch', 'from_model', 'to_model',
                         'from_model_range_plus_target']]

ground_truth = read_ground_truth('')
```
