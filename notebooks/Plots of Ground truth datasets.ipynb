```python
### Plotting synthetic dataset generated
synthetic_datasets = ['../output/timeseries_created_1576000391.csv'] # timeseries_created_1576000391_indicators.csv']
for dataset in synthetic_datasets:
    # Load synthetic set (before addition of technical indicators)
    df = pd.read_csv(base_path + dataset, sep=',')
    df.set_index('n_row', drop=True, inplace=True)
    df = df[['ts']][:1000000]
    
    # Generated series
    fig = plt.figure(figsize=(15,10))
    ax = fig.add_subplot(2, 1, 1)
    ax.set_xlim(0,1000000)
    ax.set_ylim(0,7000000)
    plot, = ax.plot(df, color='blue', lw=2)
    fig.savefig('/mnt/c/Users/suare/OneDrive/Escritorio/'+ dataset.split('/')[-1].replace('.csv', '.pdf'))
    fig
    
    # Generated series in logarithmic scale
    fig = plt.figure(figsize=(15,10))
    ax = fig.add_subplot(2, 1, 1)
    plot, = ax.plot(df, color='blue', lw=2)
    ax.set_xlim(0,1000000)
    ax.set_yscale('log')
    fig.savefig('/mnt/c/Users/suare/OneDrive/Escritorio/logarithmic'+ dataset.split('/')[-1].replace('.csv', '.pdf'))
    fig
```
